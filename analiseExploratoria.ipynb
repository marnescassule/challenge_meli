```python
import pandas as pd
```

```python
arquivos = ['arquivo1.csv', 'arquivo2.csv', 'arquivo3.csv', 'arquivo4.csv']
```

```python
dfs= []
```

```python
for arquivo in arquivos:
    df_temp = pd.read_csv(arquivo)
    dfs.append(df_temp)
```

```python
df_final = pd.concat(dfs, ignore_index=True)
```
